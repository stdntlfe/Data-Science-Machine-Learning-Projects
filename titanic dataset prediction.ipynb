```python
train["Name"]=train["Name"].cat.codes
train["Sex"]=train["Sex"].cat.codes
train["Cabin"]=train["Cabin"].cat.codes
train["Embarked"]=train["Embarked"].cat.codes
train["Ticket"]=train["Ticket"].cat.codes
test["Name"]=test["Name"].cat.codes
test["Sex"]=test["Sex"].cat.codes
test["Cabin"]=test["Cabin"].cat.codes
test["Embarked"]=test["Embarked"].cat.codes
test["Ticket"]=test["Ticket"].cat.codes
test["Age"]=test["Age"].cat.codes
test["Fare"]=test["Fare"].cat.codes
train["Age"]=train["Age"].cat.codes
train["Fare"]=train["Fare"].cat.codes
```

```python
y_train = train['Survived'].ravel()
train = train.drop(['Survived'], axis=1)
x_train = train.values # Creates an array of the train data
x_test = test.values
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#dependent ='Survived'
#independent=[x for x in train.columns if x not in['ID',dependent]]
```

```python
#initialize the Algorithm
model=DecisionTreeClassifier(max_depth=10,min_samples_leaf=100,max_features='sqrt')
```

```python
#fit the algorithm
model.fit(x_train,y_train)
```

```python
train_pred=model.predict(x_train)
test_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#determine the train errorore
acc_train=accuracy_score(y_train,train_pred)
```

```python
print("train accuracy %f"%acc_train)
```
